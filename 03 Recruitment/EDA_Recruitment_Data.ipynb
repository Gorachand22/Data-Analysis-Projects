```python
def find_outlier(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers
```

```python
def univariate_num(col, bin = 20):
    print(col.describe()) # summary
    print()

    col.plot(kind='hist',bins=bin) # histogram
    plt.title('histogram - {}'.format(col.name))
    plt.xlabel(col.name)
    plt.ylabel('Frequency')
    plt.show()
    print()
    
    col.plot(kind='kde') # density plot
    plt.title('density plot - {}'.format(col.name))
    plt.xlabel(col.name)
    plt.ylabel('Frequency')
    plt.show()

    print()

    col.plot(kind='box', vert=False) # boxplot
    plt.title('Box Plot - {}'.format(col.name))
    plt.xlabel(col.name)
    plt.ylabel('Frequency')
    plt.show()

    print()

    print("Skewness: ", col.skew())
    print("Kurtosis: ", col.kurt()+3)

    print()

    return
```

```python
def univariate_cat(col, color= ['r', 'b', 'g', 'y', 'm', 'c', 'k'], figsize=(10, 6)):
    print("valuecounts",col.value_counts())
    print()
    print("valuecounts (normalize)",col.value_counts(normalize=True))
    print()
    col.value_counts().plot(kind='bar', 
                            rot=45,
                            color = color,
                            figsize=figsize)
    plt.title('Bar plot - {}'.format(col.name))
    plt.xlabel(col.name)
    plt.ylabel('Frequency')
    plt.show()

    print()

    col.value_counts().plot(kind='pie',
                            autopct='%0.1f%%',
                            colors = color,
                            figsize=figsize)
    plt.title('Pie plot - {}'.format(col.name))
    plt.xlabel(col.name)
    plt.ylabel('Frequency')
    plt.show()
    print('\n'*4)
    return
```

```python
data1 = data.copy()
Id = data1.index
data1.insert(0, 'Id', Id)
data_pivot = pd.pivot_table(
        data1[['Id','recruiting_source', 'attrition']],
        index=['recruiting_source'],
        columns=["attrition"],
        aggfunc='count',
        fill_value=0).reset_index()

data_pivot
```

```python
def bivarite_cat_cat(data, target_column, figsize=(10, 6)):
    data1 = data.copy()
    Id = data1.index
    data1.insert(0, 'CandidateId', Id)
    data_pivot = pd.pivot_table(
        data1[['CandidateId', target_column, 'attrition']],
        index=[target_column],
        columns=["attrition"],
        aggfunc='count',
        fill_value=0).reset_index()

    data_pivot.columns = [target_column, 'not_leaved', 'leaved']

    data_pivot['candidates'] = data_pivot['not_leaved']+data_pivot['leaved']

    data_pivot['stayed_ratio'] = (data_pivot['not_leaved']/(data_pivot['candidates'])*100).round(2)
    data_pivot['left_ratio'] = (data_pivot['leaved']/(data_pivot['candidates'])*100).round(2)

    print(data_pivot.to_markdown(), end = '\n\n')
    
    ax = pd.crosstab(data1['attrition'], data1[target_column]).plot(kind='bar', 
                                                    figsize=figsize,
                                                    rot = 0,
                                                    legend = True,
                                                    title = 'Bar plot - {} vs {}'.format(data1['attrition'].name, data1[target_column].name),
                                                    xlabel = data1['attrition'].name,
                                                    ylabel = 'Count'
                                                    )
    for p in ax.patches:
        ax.annotate('{}'.format(p.get_height()), (p.get_x()+0.1, p.get_height()+10))

    plt.show()
    print('\n'*4)
    return
```

```python
for col in categorical_cols:
    if col != 'attrition':
        bivarite_cat_cat(data, col)
```
