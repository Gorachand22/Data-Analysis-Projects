```python
def univariate_num(col, bin = 20):
    print(col.describe()) # summary
    print()

    col.plot(kind='hist',bins=bin) # histogram
    plt.title('histogram - {}'.format(col.name))
    plt.xlabel(col.name)
    plt.ylabel('Frequency')
    plt.show()
    print()
    
    col.plot(kind='kde') # density plot
    plt.title('density plot - {}'.format(col.name))
    plt.xlabel(col.name)
    plt.ylabel('Frequency')
    plt.show()

    print()

    col.plot(kind='box', vert=False) # boxplot
    plt.title('Box Plot - {}'.format(col.name))
    plt.xlabel(col.name)
    plt.ylabel('Frequency')
    plt.show()

    print()

    print("Skewness: ", col.skew())
    print("Kurtosis: ", col.kurt()+3)

    print()

    return 
```

```python
def univariate_cat(col, color= ['r', 'b', 'g', 'y', 'm', 'c', 'k'], figsize=(10, 6)):
    print("valuecounts",col.value_counts())
    print()
    print("valuecounts (normalize)",col.value_counts(normalize=True))
    print()
    col.value_counts().plot(kind='bar', 
                            rot=45,
                            color = color,
                            figsize=figsize)
    plt.title('Bar plot - {}'.format(col.name))
    plt.xlabel(col.name)
    plt.ylabel('Frequency')
    plt.show()

    print()

    col.value_counts().plot(kind='pie',
                            autopct='%0.1f%%',
                            colors = color,
                            figsize=figsize)
    plt.title('Pie plot - {}'.format(col.name))
    plt.xlabel(col.name)
    plt.ylabel('Frequency')
    plt.show()

    return
```

```python
def find_outliers(column):
    # Calculate the first quartile (Q1)
    Q1 = column.quantile(0.25)
    # Calculate the third quartile (Q3)
    Q3 = column.quantile(0.75)
    # Calculate the interquartile range (IQR)
    IQR = Q3 - Q1
    # Define the lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Find outliers
    outliers = column[(column < lower_bound) | (column > upper_bound)]
    return outliers
```

            Categorical - Categorical:

```python
def bivarite_cat_cat(data, target_column, figsize=(10, 6)):
    data_pivot = pd.pivot_table(
        data[['PassengerId', target_column, 'Survived']],
        index=[target_column],
        columns=["Survived"],
        aggfunc='count',
        fill_value=0).reset_index()

    data_pivot.columns = [target_column, 'not_survived', 'survived']

    data_pivot['passengers'] = data_pivot['not_survived']+data_pivot['survived']

    data_pivot['survival_rate'] = (data_pivot['survived']/data_pivot['passengers'])*100

    print(data_pivot.to_markdown(), end = '\n\n')
    
    ax = pd.crosstab(data['Survived'], data[target_column]).plot(kind='bar', 
                                                    figsize=figsize,
                                                    rot = 0,
                                                    legend = True,
                                                    title = 'Bar plot - {} vs {}'.format(data['Survived'].name, data[target_column].name),
                                                    xlabel = data['Survived'].name,
                                                    ylabel = 'Count'
                                                    )
    for p in ax.patches:
        ax.annotate('{}'.format(p.get_height()), (p.get_x()+0.1, p.get_height()+10))

    plt.show()
    return
```

```python
import re
def extract_name_info(name):
    surname = re.sub(r'(, \w+\..*)','', name).strip()
    title = re.findall(r'(\w+)\.', name)[0].strip()
    a = re.findall(r'\w+\.(.*)', name)[0].strip()
    try:
        firstname = a.split()[0].strip().replace('(', '').replace(')', '')
    except:
        firstname = ''
    
    try:
        lastname = a.split()[1].strip().replace('(', '').replace(')', '')
    except:
        lastname = ''

    return surname, title, firstname, lastname
```

```python
def multivarite_cat_cat(data, target_column1, target_column2, figsize=(10, 6)):
    data_pivot = pd.pivot_table(
        data[['PassengerId', target_column1,target_column2, 'Survived']],
        index=[target_column1, target_column2],
        columns=["Survived"],
        aggfunc='count',
        fill_value=0).reset_index()

    data_pivot.columns = [target_column1, target_column2, 'not_survived', 'survived']

    data_pivot['passengers'] = data_pivot['not_survived']+data_pivot['survived']

    data_pivot['survival_rate'] = (data_pivot['survived']/data_pivot['passengers'])*100

    print(data_pivot.to_markdown(), end = '\n\n')
    
    X= data_pivot[['Sex', 'Pclass', 'survival_rate']]

    plt.figure(figsize=figsize)
    ax = sns .barplot(x=target_column1, y='survival_rate', hue=target_column2, data=X)
    for p in ax.patches:
        # add round figur
        ax.annotate(np.round(p.get_height(),decimals=2), 
                    (p.get_x()+p.get_width()/2., p.get_height()), 
                    ha='center', 
                    va='center', 
                    xytext=(0, 10), 
                    textcoords='offset points')

    plt.show()
    return
```
